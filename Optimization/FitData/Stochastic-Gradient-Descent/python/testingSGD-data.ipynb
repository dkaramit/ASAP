I wrote SGD in the form of a module, just to be able to add more algorithms and test it easiy.

This is how it works

```python
import numpy as np


import matplotlib
matplotlib.use('nbAgg')
import matplotlib.pyplot as plt
```

```python
import SGD
```

```python
data_in=[]
data_out=[]
xmin=-5
xmax=2
for i in range(5000):
    x=np.random.rand()*(xmax-xmin)+xmin
    
    data_in.append( [x])
    data_out.append( [2*x+3])
```

```python
class modelFunc(SGD.modelBase):
    '''class derived from modelBase'''
    
    #you don't really need this, but it remids you how to declare instances (also it allows you to make add new attributes)
    def __init__(self,dimensions,w0):
        SGD.modelBase.__init__(self,dimensions,w0)
        
        
    def __call__(self):
        self.signal[0]=self.input[0]*self.w[0]+ self.w[1]
        
        
    def derivative_w(self,i):
        if i==0:
            self.dsdw[0]=self.input[0]
        if i==1:
            self.dsdw[0]=1
```

```python
#the squared error and its derivative
def Q_i(signal, target):
    return (signal-target)**2


def dQds_i(signal, target):
    return 2*(signal-target)

        
#declare the instances
model=modelFunc([1,1],[1,0.2])
Q=SGD.lossFunc(data_in,data_out,Q_i,dQds_i,model)
```

```python
# sgd=SGD.VanillaSGD(Q,alpha=1e-2)
# sgd=SGD.RMSpropSGD(Q,gamma=1-1e-2,epsilon=1e-5,alpha=1e-2)
# sgd=SGD.AdaDeltaSGD(Q,gamma=0.99,epsilon=1e-5,alpha=1)
# sgd=SGD.AdamSGD(Q,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-2)
# sgd=SGD.AdaMaxSGD(Q,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-2)
sgd=SGD.NAdamSGD(Q,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-2)
```

```python
w=sgd.run(abs_tol=1e-5, rel_tol=1e-5, step_break=250,max_step=5000)

w,len(sgd.steps)
```

```python
fig=plt.figure(figsize=(8,6))
fig.subplots_adjust(bottom=0.15, left=0.15, top = 0.9, right=0.9,wspace=0.1,hspace=0)
#=============================================================================#


sub = fig.add_subplot(211)
sub.get_xaxis().set_visible(False)

sub.plot(np.array(sgd.steps)[:,0],c='xkcd:black')
sub.plot(np.array(sgd.steps)[:,1],c='xkcd:red')

sub.set_ylabel('x')
# sub.set_xscale('log')
sub.set_yscale('linear')


sub = fig.add_subplot(212)
meanQ=[]
steps=[]
_s=0
skip=100
for w in sgd.steps[::skip]:
    model.w=w
    mQ=0
    _r=range(len(data_in))
    mC=50
    for _ in range(mC): 
        i=np.random.choice(_r)
        model.setInput(data_in[i])
        model()
        mQ+=np.abs( Q(data_out[i]) )
        
    if mQ==0:
        mQ=1e-50
    meanQ.append(np.log10(mQ/mC))
    steps.append(_s)
    _s+=skip
    
sub.plot(steps,meanQ,c='xkcd:black')

sub.set_ylabel(r'$log_{10} \langle |Q| \rangle $')

# sub.set_xscale('log')
sub.set_yscale('linear')

sub.set_xlabel('Step')

plt.show()
```
