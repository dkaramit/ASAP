```python
steps=np.array([_.split(' ') for _ in _cpp.split("\n")[:-3]],dtype=np.float)

x,minimum=[np.array(_.split(' '),dtype=np.float) for _ in _cpp.split("\n")[-3:-1]]
len(steps),x,minimum
```
