# Euler in *C++*

This notebook is here in order to plot results from the C++ implementation
of Euler.

```python
import numpy as np

import matplotlib
#matplotlib.use('WebAgg')
#matplotlib.use('Qt4Cairo')
#matplotlib.use('Qt5Cairo')
matplotlib.use('nbAgg')
import matplotlib.pyplot as plt

plt.rcParams['font.family']='serif'
plt.rcParams['font.size']=10
plt.rcParams['mathtext.fontset']='stixsans'
```

```python
%%sh
#compile and run Euler in c++
g++ Euler.cpp -o Eu
./Eu
rm Eu
```

```python
y1=np.loadtxt('data/y1.dat')
y2=np.loadtxt('data/y2.dat')
t=np.loadtxt('data/t.dat')
```

```python
%%sh
#in data I have Euler in python. using the same parameters it should give the identical resuls.

cd data/
python3 Euler.py
```

```python

y1_py=np.loadtxt('data/y1_py.dat')
y2_py=np.loadtxt('data/y2_py.dat')
t_py=np.loadtxt('data/t_py.dat')
```

```python
fig=plt.figure(figsize=(8,4.5))
fig.subplots_adjust(bottom=0.15, left=0.1, top = 0.9, right=0.9,wspace=0.0,hspace=0.0)
fig.suptitle('')


sub = fig.add_subplot(111)
sub.plot(t,y1,linestyle='-',linewidth=2,c='xkcd:red',alpha=0.5,label=r'$y_1(t)$ c++')
sub.plot(t,y2,linestyle='-',linewidth=2,c='xkcd:black',alpha=0.5,label=r'$y_2(t)$ c++')

sub.plot(t_py,y1_py,linestyle='--',linewidth=3,c='xkcd:red',label=r'$y_1(t)$ py')
sub.plot(t_py,y2_py,linestyle='--',linewidth=3,c='xkcd:black',label=r'$y_1(t)$ py')

sub.legend()

plt.show()
```
